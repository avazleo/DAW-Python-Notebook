## 1.- Factorial de un n√∫mero

Crea una funci√≥n factorial(n) que devuelva el factorial de n usando recursividad

```python
def factorial(n):
    if n == 0:
        return 1
    return n * factorial(n - 1)


def main():
    n = int(input('Enter a number: '))
    print(f"The factorial of {n} is {factorial(n)}")


if __name__ == '__main__':
    main()
```

## 2 ‚Äì Suma de los primeros n n√∫meros
Crea una funci√≥n suma_n(n) que calcule la suma de los primeros n n√∫meros naturales usando
recursividad.

```python
def suma_n(n):
    if n == 0:
        return 0
    return n + suma_n(n - 1)


def main():
    n = int(input('Enter a number: '))
    print(f"The sum of the first {n} numbers is {suma_n(n)}")


if __name__ == '__main__':
    main()
```

## 3 ‚Äì Potencia de un n√∫mero
Crea una funci√≥n potencia(base, exponente) que calcule base ** exponente sin usar el operador **,
solo recursi√≥n.

```python
def potencia(base, exponente):
    if exponente == 0:
        return 1
    return base * potencia(base, exponente - 1)

def main():
    base = int(input('Enter the base: '))
    exponente = int(input('Enter the exponent: '))
    print(f"{base} raised to the power of {exponente} is {potencia(base, exponente)}")

if __name__ == '__main__':
    main()
```

## 4 ‚Äì Invertir una cadena
Crea una funci√≥n invertir(cadena) que devuelva la cadena invertida de forma recursiva.

```python
def invertir(cadena):
    if len(cadena) <= 1:
        return cadena
    return invertir(cadena[1:]) + cadena[0]

def main():
    cadena = input('Enter a string: ')
    print(f"The reversed string is: {invertir(cadena)}")

if __name__ == '__main__':
    main()
```

## 5 ‚Äì Suma de elementos de una lista
Crea una funci√≥n suma_lista(lista) que calcule la suma de todos los elementos de una lista de forma recursiva.

```python
def suma_lista(lista):
    if not lista:
        return 0
    return lista[0] + suma_lista(lista[1:])

def main():
    # Example usage:
    my_list = [1, 2, 3, 4, 5]
    print(f"The sum of the list {my_list} elements is: {suma_lista(my_list)}")

if __name__ == '__main__':
    main()
```

## 6 ‚Äì Contar elementos en una lista
Crea una funci√≥n contar_elementos(lista) que devuelva cu√°ntos elementos tiene la lista sin usar len().

```python
def contar_elementos(lista):
    if not lista:
        return 0
    return 1 + contar_elementos(lista[1:])

def main():
    # Example usage:
    my_list = [10, 20, 30, 40, 50, 60]
    print(f"The number of elements in the list {my_list} is: {contar_elementos(my_list)}")

if __name__ == '__main__':
    main()
```

## 7 ‚Äì N-√©simo n√∫mero de Fibonacci
Crea una funci√≥n fibonacci(n) que devuelva el t√©rmino n-√©simo de la secuencia de Fibonacci.

```python
def fibonacci(n):
    if n <= 1:
        return n
    else:
        return fibonacci(n-1) + fibonacci(n-2)

def main():
    n = int(input("Enter a number: "))
    if n < 0:
        print("Fibonacci sequence is not defined for negative numbers")
    else:
        print(f"The {n}-th Fibonacci number is: {fibonacci(n)}")

if __name__ == '__main__':
    main()
```

## 8 ‚Äì Buscar un elemento en una lista
Crea una funci√≥n buscar(lista, valor) que devuelva True si el valor est√° en la lista, o False si no.

```python
def buscar(lista, valor):
    if not lista:
        return False
    if lista[0] == valor:
        return True
    return buscar(lista[1:], valor)

def main():
    my_list = [10, 20, 30, 40, 50]
    value_to_find = 30
    if buscar(my_list, value_to_find):
        print(f"{value_to_find} is in the list.")
    else:
        print(f"{value_to_find} is not in the list.")

    value_to_find = 60
    if buscar(my_list, value_to_find):
        print(f"{value_to_find} is in the list.")
    else:
        print(f"{value_to_find} is not in the list.")

if __name__ == '__main__':
    main()
```

## 9 ‚Äì Contar d√≠gitos de un n√∫mero
Crea una funci√≥n contar_digitos(n) que devuelva cu√°ntos d√≠gitos tiene un n√∫mero entero positivo.

```python
def contar_digitos(n):
    if n < 10:
        return 1
    else:
        return 1 + contar_digitos(n // 10)

def main():
    n = int(input("Enter a positive integer: "))
    if n < 0:
        print("Please enter a positive integer.")
    else:
        print(f"The number of digits in {n} is: {contar_digitos(n)}")

if __name__ == '__main__':
    main()
```

## 10 ‚Äì Torres de Hanoi
Crea una funci√≥n hanoi(n, origen, auxiliar, destino) que muestre los pasos para mover n discos desde la torre de origen hasta la de destino, usando la torre auxiliar.

```python
def hanoi(n, origen, auxiliar, destino):
    if n == 1:
        print(f"Move disc {origen} ‚Üí {destino}")
        return

    hanoi(n - 1, origen, destino, auxiliar)
    print(f"Move disc {origen} ‚Üí {destino}")
    hanoi(n - 1, auxiliar, origen, destino)

def main():
  # Example: Move 3 disks from tower A to C using tower B as an auxiliary.
  hanoi(3, 'A', 'B', 'C')

if __name__ == '__main__':
    main()
```

Alternativa para entenderlo mejor.

```python
def mostrar_torres(A, B, C):
    """It shows the current status of the three towers."""
    print(f"A: {A}")
    print(f"B: {B}")
    print(f"C: {C}")
    print("-" * 20)

def hanoi(n, origen, auxiliar, destino, torres):
    if n == 1:
        disco = torres[origen].pop()       # Remove the top disk from the source
        torres[destino].append(disco)      # Place it in the destination
        print(f"Move disc {disco} from {origen} ‚Üí {destino}")
        mostrar_torres(torres['A'], torres['B'], torres['C'])
        return

    # Move n-1 disks from source to auxiliary
    hanoi(n - 1, origen, destino, auxiliar, torres)

    # Move the largest disc to the destination
    disco = torres[origen].pop()
    torres[destino].append(disco)
    print(f"Mueve disco {disco} de {origen} ‚Üí {destino}")
    mostrar_torres(torres['A'], torres['B'], torres['C'])

    # Move the n-1 disks from the auxiliary to the destination
    hanoi(n - 1, auxiliar, origen, destino, torres)


# --- Example of use ---
n = 3  # number of disks
torres = {
    'A': list(range(n, 0, -1)),  # origin
    'B': [],                     # assistant
    'C': []                      # destination
}

print("Initial state:")
mostrar_torres(torres['A'], torres['B'], torres['C'])

hanoi(n, 'A', 'B', 'C', torres)
print("Towers completed! üéâ")
```
